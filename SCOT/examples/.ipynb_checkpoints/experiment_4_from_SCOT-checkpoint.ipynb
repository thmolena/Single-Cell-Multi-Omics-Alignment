```python
# eval function

import numpy as np
import random, math, os, sys
import matplotlib.pyplot as plt
from sklearn.preprocessing import normalize
from sklearn.metrics import roc_auc_score, silhouette_samples
from sklearn.decomposition import PCA
from sklearn.neighbors import KNeighborsClassifier

def calc_frac_idx(x1_mat,x2_mat):
    """
    Returns fraction closer than true match for each sample (as an array)
    """
    fracs = []
    x = []
    nsamp = x1_mat.shape[0]
    rank=0
    for row_idx in range(nsamp):
        euc_dist = np.sqrt(np.sum(np.square(np.subtract(x1_mat[row_idx,:], x2_mat)), axis=1))
        true_nbr = euc_dist[row_idx]
        sort_euc_dist = sorted(euc_dist)
        rank =sort_euc_dist.index(true_nbr)
        frac = float(rank)/(nsamp -1)

        fracs.append(frac)
        x.append(row_idx+1)

    return fracs,x

def calc_domainAveraged_FOSCTTM(x1_mat, x2_mat):
    """
    Outputs average FOSCTTM measure (averaged over both domains)
    Get the fraction matched for all data points in both directions
    Averages the fractions in both directions for each data point
    """
    
    fracs1,xs = calc_frac_idx(x1_mat, x2_mat)
    fracs2,xs = calc_frac_idx(x2_mat, x1_mat)
    fracs = []
    for i in range(len(fracs1)):
        fracs.append((fracs1[i]+fracs2[i])/2)  
    return fracs

def calc_sil(x1_mat,x2_mat,x1_lab,x2_lab):
    """
    Returns silhouette score for datasets with cell clusters
    """
    sil = []
    sil_d0 = []
    sil_d3 = []
    sil_d7 = []
    sil_d11 = []
    sil_npc = []

    x = np.concatenate((x1_mat,x2_mat))
    lab = np.concatenate((x1_lab,x2_lab))

    sil_score = silhouette_samples(x,lab)

    nsamp = x.shape[0]
    for i in range(nsamp):
        if(lab[i]==1):
            sil_d0.append(sil_score[i])
        elif(lab[i]==2):
            sil_d3.append(sil_score[i])
        elif(lab[i]==3):
            sil_d7.append(sil_score[i])
        elif(lab[i]==4):
            sil_d11.append(sil_score[i])
        elif(lab[i]==5):
            sil_npc.append(sil_score[i])

    avg = np.mean(sil_score)
    d0 = sum(sil_d0)/len(sil_d0)
    d3 = sum(sil_d3)/len(sil_d3)
    d7 = sum(sil_d7)/len(sil_d7)
    d11 = sum(sil_d11)/len(sil_d11)
    npc = sum(sil_npc)/len(sil_npc)

    return avg,d0,d3,d7,d11,npc

def binarize_labels(label,x):
    """
    Helper function for calc_auc
    """
    bin_lab = np.array([1] * len(x))
    idx = np.where(x == label)

    bin_lab[idx] = 0
    return bin_lab


def calc_auc(x1_mat, x2_mat, x1_lab, x2_lab):
    """
    calculate avg. ROC AUC scores for transformed data when there are >=2 number of clusters.
    """

    nsamp = x1_mat.shape[0]

    auc = []
    auc_d0 = []
    auc_d3 = []
    auc_d7 = []
    auc_d11 = []
    auc_npc = []

    for row_idx in range(nsamp):
        euc_dist = np.sqrt(np.sum(np.square(np.subtract(x1_mat[row_idx,:], x2_mat)), axis=1))
        y_scores = euc_dist
        y_true = binarize_labels(x1_lab[row_idx],x2_lab)

        auc_score = roc_auc_score(y_true, y_scores)
        auc.append(auc_score)
        
        if(x1_lab[row_idx]==0):
            auc_d0.append(auc_score)
        elif(x1_lab[row_idx]==1):
            auc_d3.append(auc_score)
        elif(x1_lab[row_idx]==2):
            auc_d7.append(auc_score)
        elif(x1_lab[row_idx]==3):
            auc_d11.append(auc_score)
        elif(x1_lab[row_idx]==4):
            auc_npc.append(auc_score)

    avg = sum(auc)/len(auc)
    d0 = sum(auc_d0)/len(auc_d0)
    d3 = sum(auc_d3)/len(auc_d3)
    d7 = sum(auc_d7)/len(auc_d7)
    d11 = sum(auc_d11)/len(auc_d11)
    npc = sum(auc_npc)/len(auc_npc)

    return avg,d0,d3,d7,d11,npc

def transfer_accuracy(domain1, domain2, type1, type2, n):
    """
    Metric from UnionCom: "Label Transfer Accuracy"
    """
    knn = KNeighborsClassifier(n_neighbors=n)
    knn.fit(domain2, type2)
    type1_predict = knn.predict(domain1)
    np.savetxt("type1_predict.txt", type1_predict)
    count = 0
    for label1, label2 in zip(type1_predict, type1):
        if label1 == label2:
            count += 1
    return count / len(type1)
```

```python
from sklearn.manifold import MDS
import numpy as np
import matplotlib.pylab as pl
import torch

import ot
from ot.gromov import gromov_wasserstein2

rng = np.random.RandomState(42)


def get_sbm(n, nc, ratio, P):
    nbpc = np.round(n * ratio).astype(int)
    n = np.sum(nbpc)
    C = np.zeros((n, n))
    for c1 in range(nc):
        for c2 in range(c1 + 1):
            if c1 == c2:
                for i in range(np.sum(nbpc[:c1]), np.sum(nbpc[:c1 + 1])):
                    for j in range(np.sum(nbpc[:c2]), i):
                        if rng.rand() <= P[c1, c2]:
                            C[i, j] = 1
            else:
                for i in range(np.sum(nbpc[:c1]), np.sum(nbpc[:c1 + 1])):
                    for j in range(np.sum(nbpc[:c2]), np.sum(nbpc[:c2 + 1])):
                        if rng.rand() <= P[c1, c2]:
                            C[i, j] = 1

    return C + C.T


n = 100
nc = 3
ratio = np.array([.5, .3, .2])
P = np.array(0.6 * np.eye(3) + 0.05 * np.ones((3, 3)))
C1 = get_sbm(n, nc, ratio, P)

# get 2d position for nodes
x1 = MDS(dissimilarity='precomputed', random_state=0).fit_transform(1 - C1)


C0 = np.eye(3) 

def min_weight_gw(C1, C2, a2, nb_iter_max=100, lr=1e-2):
    """ solve min_a GW(C1,C2,a, a2) by gradient descent"""

    # use pyTorch for our data
    C1_torch = torch.tensor(C1)
    C2_torch = torch.tensor(C2)

    a0 = rng.rand(C1.shape[0])  # random_init
    a0 /= a0.sum()  # on simplex
    a1_torch = torch.tensor(a0).requires_grad_(True)
    a2_torch = torch.tensor(a2)

    loss_iter = []

    for i in range(nb_iter_max):

        loss = gromov_wasserstein2(C1_torch, C2_torch, a1_torch, a2_torch)

        loss_iter.append(loss.clone().detach().cpu().numpy())
        loss.backward()

        #print("{:03d} | {}".format(i, loss_iter[-1]))

        # performs a step of projected gradient descent
        with torch.no_grad():
            grad = a1_torch.grad
            a1_torch -= grad * lr   # step
            a1_torch.grad.zero_()
            a1_torch.data = ot.utils.proj_simplex(a1_torch)

    a1 = a1_torch.clone().detach().cpu().numpy()

    return a1, loss_iter


a0_est, loss_iter0 = min_weight_gw(C0, C1, ot.unif(n), nb_iter_max=100, lr=1e-2)

pl.figure(2)
pl.plot(loss_iter0)
pl.title("Loss along iterations")
a0_est
```

```python
import torch.optim as optim

def min_weight_gw(C1, C2, a2, nb_iter_max=100, lr=1e-2):
    """ solve min_a GW(C1,C2,a, a2) by gradient descent"""

    # use PyTorch for our data
    C1_torch = torch.tensor(C1)
    C2_torch = torch.tensor(C2)

    a0 = rng.rand(C1.shape[0])  # random_init
    a0 /= a0.sum()  # on simplex
    a1_torch = torch.tensor(a0).requires_grad_(True)
    a2_torch = torch.tensor(a2)

    # Define the SGD optimizer
    optimizer = optim.SGD([a1_torch], lr=lr)

    loss_iter = []

    for i in range(nb_iter_max):

        loss = gromov_wasserstein2(C1_torch, C2_torch, a1_torch, a2_torch)

        loss_iter.append(loss.clone().detach().cpu().numpy())

        # Backpropagate the loss and update the weights
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        # Project the weights onto the simplex
        with torch.no_grad():
            a1_torch.data = ot.utils.proj_simplex(a1_torch)

    a1 = a1_torch.clone().detach().cpu().numpy()

    return a1, loss_iter

a0_est, loss_iter0 = min_weight_gw(C0, C1, ot.unif(n), nb_iter_max=100, lr=1e-2)

pl.figure(2)
pl.plot(loss_iter0)
pl.title("Loss along iterations")
a0_est
```

```python
import torch
import ot
import torch.nn.functional as F
from torch import nn, optim
import pytorch_metric_learning.losses as loss
import pytorch_metric_learning.miners as miners
import matplotlib.pylab as pl
from sklearn.cluster import KMeans


def min_gromov_wasserstein_distance(C1, C2, p, q, nb_iter_max=10, lr=1e-2):
    
    loss_iter = []
    optimizer = torch.optim.SGD([C1,C2], lr=0.02)

    for i in range(nb_iter_max):
        optimizer.zero_grad()
        loss = ot.gromov.entropic_gromov_wasserstein2(C1, C2, p, q, loss_fun="square_loss", epsilon=0.1, max_iter=1000, tol=1e-09, verbose=False, log=False)
        
               
        loss_iter.append(loss.clone().detach().cpu().numpy())
        
        
        loss.backward()

        
  
        # Project the updated arrays onto the probability simplex
        C1 = ot.utils.proj_simplex(C1)
        C2 = ot.utils.proj_simplex(C2)

        # Convert the projected arrays back to PyTorch tensors
        C1 = torch.tensor(C1, requires_grad=True)
        C2 = torch.tensor(C2, requires_grad=True)
        
        optimizer.step()
  

    # Convert the final tensors back to numpy arrays
    C1 = C1.detach().cpu().numpy()
    C2 = C2.detach().cpu().numpy()
    
    return C1, C2, loss_iter
```

```python
import warnings
warnings.filterwarnings('ignore')


import ot
import matplotlib.pylab as pl

import random
import numpy as np

n = 5
C1 = np.zeros((n, n))

for i in range(n):
    for j in range(i+1, n):
        s1 = ''.join(random.choices(['0', '1'], k=5))
        s2 = ''.join(random.choices(['0', '1'], k=5))
        dist = sum([s1[k] != s2[k] for k in range(5)])
        C1[i,j] = dist
        C1[j,i] = dist

print("Distance Matrix in Hamming Metric Space:")
print(C1.shape)



n = 5
C2 = np.zeros((n, n))

for i in range(n):
    for j in range(i+1, n):
        s1 = ''.join(random.choices(['A', 'C', 'G', 'T'], k=5))
        s2 = ''.join(random.choices(['A', 'C', 'G', 'T'], k=5))
        m = len(s1)
        n_ = len(s2)
        dp = [[0 for _ in range(n_+1)] for _ in range(m+1)]
        for i_ in range(m+1):
            dp[i_][0] = i_
        for j_ in range(n_+1):
            dp[0][j_] = j_
        for i_ in range(1, m+1):
            for j_ in range(1, n_+1):
                if s1[i_-1] == s2[j_-1]:
                    dp[i_][j_] = dp[i_-1][j_-1]
                else:
                    dp[i_][j_] = 1 + min(dp[i_-1][j_], dp[i_][j_-1], dp[i_-1][j_-1])
        C2[i,j] = dp[m][n_]
        C2[j,i] = dp[m][n_]

print("Distance Matrix in Levenshtein Metric Space:")
print(C2.shape)



C1 = torch.tensor(C1, dtype=torch.float64, requires_grad=True)
C2 = torch.tensor(C2, dtype=torch.float64, requires_grad=True)




p = ot.unif(C1.shape[0])
p = torch.tensor(p)
q = ot.unif(C2.shape[0])
q = torch.tensor(q)
C1_opt, C2_opt, loss_iter = min_gromov_wasserstein_distance(C1, C2, p, q, nb_iter_max=100, lr=1e-2)
    

pl.figure(2)
pl.plot(loss_iter)
pl.title("Loss along iterations")
```

```python
import warnings
warnings.filterwarnings('ignore')

import torch
import ot
import torch.nn.functional as F
from torch import nn, optim
import pytorch_metric_learning.losses as loss
import pytorch_metric_learning.miners as miners
import matplotlib.pylab as pl
from sklearn.cluster import KMeans


C1 = torch.randn(100, 100, requires_grad=True)
C2 = torch.randn(100, 100, requires_grad=True)
p = ot.unif(C1.shape[0])
p = torch.tensor(p)
q = ot.unif(C2.shape[0])
q = torch.tensor(q)



# Define model
model = torch.nn.Linear(C1.shape[0], 5)
 

C1_concat = torch.cat((C1, C1), dim=0).detach()
C1_concat.requires_grad=True

C2_concat = concatenated_tensor = torch.cat((C2, C2), dim=0)

# # perform k-means clustering with k=2
# C2_kmeans = KMeans(n_clusters=10)
# C2_kmeans.fit(C2_concat.detach().numpy())

# # get the cluster labels and convert them to a tensor
# C2_labels = C2_kmeans.labels_
# C2_labels = torch.from_numpy(C2_labels).type(torch.int64)


# Define miner
miner = miners.MultiSimilarityMiner()

# Define loss function
loss_fn = loss.ProxyAnchorLoss(num_classes=120, embedding_size=5)

# make a copy of A

C1_fixed = C1.clone().detach()
C1_fixed.requires_grad=True

C2_fixed = C2.clone().detach()
C2_fixed.requires_grad=True


# Train model
loss_iter = []


# Define optimizer
optimizer = optim.SGD([
    {'params': model.parameters()},
    {'params': loss_fn.parameters()},
    {'params': [C1_concat,C1,C2], 'lr': 0.001}
    # add A with its own learning rate
], lr=0.01)


def min_gromov_wasserstein_distance(C1_fixed, C2_fixed, C1, C2, p, q, nb_iter_max=10, lr=1e-2):
    
    loss_iter = []
    
    for i in range(nb_iter_max):
    
        optimizer.zero_grad()
        d_gw_loss = ot.gromov_wasserstein2(C1,C2,p,q)
        
        # Project the updated arrays onto the probability simplex
        C1 = ot.utils.proj_simplex(C1)
        C2 = ot.utils.proj_simplex(C2)

        # Convert the projected arrays back to PyTorch tensors
        C1 = torch.tensor(C1, requires_grad=True)
        C2 = torch.tensor(C2, requires_grad=True)

        C1_concat = torch.cat((C1, C1_fixed), dim=0).detach()
        C1_concat.requires_grad=True
    
        C1_embeddings = model(C1_concat)
        
        # perform k-means clustering with k=2
        C1_kmeans = KMeans(n_clusters=10)
        C1_kmeans.fit(C1_concat.detach().numpy())

        # get the cluster labels and convert them to a tensor
        C1_labels = C1_kmeans.labels_
        C1_labels = torch.from_numpy(C1_labels).type(torch.int64)

        C1_miner_output = miner(C1_embeddings, C1_labels)
   
        C1_loss = loss_fn(C1_embeddings, C1_labels, C1_miner_output)
   
        
    
    
        # Add a penalty to the loss to encourage the output to be close to the fixed rows
        penalty = F.mse_loss(C1_concat[:C1_fixed.shape[0],:], C1_fixed)
        loss = d_gw_loss + C1_loss + penalty

        loss_iter.append(loss.clone().detach().cpu().numpy())
        loss.backward()
        
      
        
        optimizer.step()
        
        

    # Convert the final tensors back to numpy arrays
    C1 = C1.detach().cpu().numpy()
    C2 = C2.detach().cpu().numpy()
    
    return C1, C2, loss_iter


C1_opt, C2_opt, loss_iter = min_gromov_wasserstein_distance(C1_fixed, C2_fixed, C1, C2, p, q, nb_iter_max=10, lr=1e-2)

pl.figure(2)
pl.plot(loss_iter)
pl.title("Loss along iterations")
```

```python
import torch.nn.functional as F

from torch import nn, optim
import pytorch_metric_learning.losses as loss
import pytorch_metric_learning.miners as miners
import matplotlib.pylab as pl
from sklearn.cluster import KMeans

A = torch.randn(1000,1000, requires_grad=True)
print('before',A)


# perform k-means clustering with k=2
kmeans = KMeans(n_clusters=1000)
kmeans.fit(A.detach().numpy())

# get the cluster labels and convert them to a tensor
labels = kmeans.labels_
labels = torch.from_numpy(labels).type(torch.int64)


# Define model
model = torch.nn.Linear(1000, 5)
 


# Define miner
miner = miners.MultiSimilarityMiner()

# Define loss function
loss_fn = loss.ProxyAnchorLoss(num_classes=1200, embedding_size=5)

# make a copy of A
best_A = A.clone().detach()
best_A.requires_grad=True


# Define optimizer
optimizer = optim.SGD([
    {'params': model.parameters()},
    {'params': loss_fn.parameters()},
    {'params': [best_A], 'lr': 0.01},
    # add A with its own learning rate
], lr=0.1)


# Train model
loss_iter = []


for i in range(100):
    optimizer.zero_grad()
    embeddings = model(best_A)
    miner_output = miner(embeddings, labels)
   
    loss = loss_fn(embeddings, labels, miner_output)
    
    # Add a penalty to the loss to encourage the output to be close to the fixed rows
    penalty = F.mse_loss(best_A[:500,:], A[:500,:])
    loss += penalty

    
    loss_iter.append(loss.clone().detach().cpu().numpy())
    
    loss.backward()
    
    optimizer.step()  # Add this line to update the optimizer parameters
    
    # update A
    with torch.no_grad():
        best_A -= 0.01 * best_A.grad
        best_A.grad.zero_()    

print('Best A:', best_A)
pl.figure(2)
pl.plot(loss_iter)
pl.title("Loss along iterations")
```

```python
import torch.optim as optim
from pytorch_metric_learning import losses, miners, samplers, utils


# Define the dataset and labels

A = torch.randn(1000, 1000)
print('original',A[500:, :])
labels = torch.randint(low=0, high=10, size=(1000,))

# Define the samplers, miners, and loss functions
sampler = samplers.MPerClassSampler(labels, m=2, length_before_new_iter=100)
miner = miners.MultiSimilarityMiner()
loss_func = losses.ProxyAnchorLoss(num_classes=10, embedding_size=100)

# Define the model and optimizer
model = torch.nn.Linear(1000, 100)
optimizer = optim.SGD(model.parameters(), lr=0.01)

# Train the model to minimize the ProxyAnchorLoss on the last 500 rows of A
dataloader = torch.utils.data.DataLoader(A, batch_size=64, sampler=sampler, drop_last=True)
model.train()
loss_iter = []
for epoch in range(100):
    for batch in dataloader:
        optimizer.zero_grad()
        embeddings = model(batch)
        hard_pairs = miner(embeddings, labels[:len(embeddings)])
        loss = loss_func(embeddings, labels[:len(embeddings)], hard_pairs)
        

        loss_iter.append(loss.clone().detach().cpu().numpy())
        
        loss.backward()
        optimizer.step()

# Output the optimized last 500 rows of A with the minimized ProxyAnchorLoss

model.eval()
with torch.no_grad():
    new = model(A[500:, :])  # Update the last 500 rows of A with the optimized embeddings
print('new',new)
pl.figure(2)
pl.plot(loss_iter)
pl.title("Loss along iterations")
```

```python
import torch
import torch.nn.functional as F

# Generate two random distance matrices
distance_matrix_1 = torch.rand((1000, 1000))
distance_matrix_2 = torch.rand((1000, 1000))

# Define the InfoNCELoss function
def InfoNCELoss(x, y, temperature=0.5):
    x = F.normalize(x, dim=1)
    y = F.normalize(y, dim=1)
    n = x.shape[0]
    logits = torch.mm(x, y.t()) / temperature
    labels = torch.arange(n).to(x.device)
    loss = F.cross_entropy(logits, labels)
    return loss

# Calculate the loss between the two distance matrices
loss = InfoNCELoss(distance_matrix_1, distance_matrix_2)

print(f"The InfoNCELoss between the two distance matrices is {loss.item()}")
```
